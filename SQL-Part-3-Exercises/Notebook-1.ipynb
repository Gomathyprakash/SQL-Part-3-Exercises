```sql
-- Code Here:
--The title of the most tagged book, and the number of times the book has been tagged.

select top 1 b.Original_title, bt.count 
from BooksDB.dbo.Books as b
left join BooksDB.dbo.book_tags as bt
on b.book_id = bt.goodreads_book_id
where Original_title is not null 
order by bt.count desc

--The tag id for the most popular tag used for the most tagged book.

select top 1 b.original_title, bt.tag_id, bt.count as most_tagged
from BooksDB.dbo.books as b 
right join BooksDB.dbo.book_tags as bt
on b.book_id = bt.goodreads_book_id
where Original_title is not null 
order by most_tagged desc

--The tag name for the most popular tag used for the most tagged book.

select top 1 b.original_title, bt.tag_id, bt.count as most_tagged, t.tag_name
from BooksDB.dbo.books as b 
right join BooksDB.dbo.book_tags as bt
on b.book_id = bt.goodreads_book_id
inner join BooksDB.dbo.tags as t
on bt.tag_id = t.tag_id
where Original_title is not null 
order by most_tagged desc
```

**Part B:** Organize your newly joined table in a way that groups popularity based on the title based on users.

> _Hint:_ Suggest using an alias for the aggregation you will need to run on the **user\_id**

> _Minimum Desired Output:_ title and aliased column
> 
>   
> 
>

```sql
-- Code Here:

select  count(tr.user_id) as Users, b.title from 
BooksDb.dbo.to_read as tr  
left join BooksDb.dbo.Books as b 
on tr.book_id = b.book_id
where title is not null
group by title
order by Users desc
```
